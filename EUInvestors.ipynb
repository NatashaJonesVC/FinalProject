```python
#Dividing our investors into growth, early stage and pre-seed. Growth is when avg cheque size above 20M, early-stage 20-1M, pre-seed: 0-1M 
#Growth = 3 ; Earlystage = 2; Pre-Seed = 1

def numeric_tostage(x):
    if x>30:
        return 3
    elif x<1:
        return 1
    else:
        return 2
investor4['stage'] = investor4['avg_cheque'].apply(numeric_tostage)

investor4.head(10)
```

```python
#Linear regression on funds not by country

from sklearn.linear_model import LinearRegression

m2 = LinearRegression()

X1 = investor4[['aum']]
Y1 = investor4[['exits']]


# Training the model on the entire data set
m2.fit(X1,Y1)

Exit_pred = m2.predict(X1)


investor4['Exit_pred'] = Exit_pred
```

```python
# visualisation the relationship between aum and number of exits to visualise the model we have built

plt.figure(figsize= (10,8))

plt.scatter(investor4['aum'], investor4['exits'])
plt.plot(investor4['aum'], investor4["Exit_pred"], color='red') #added line 
plt.ylabel("Total Exits")
plt.xlabel("AUM ($m)")
plt.title("Relationship between AUM of Venture Capital Funds and Number of Exits");

#putting a hash for all plot saves so it does overwrite current saved copies
#plt.savefig('AUMExits.png')
```

```python
#how good is the model?

from sklearn import metrics

intercept = m2.intercept_
coefficient = m2.coef_
R2 = m2.score(X1, Y1)
MAE= metrics.mean_absolute_error(investor4['exits'], Exit_pred)
MSE = metrics.mean_squared_error(investor4['exits'], Exit_pred)
RMSE= np.sqrt(metrics.mean_squared_error(investor4['exits'], Exit_pred))


table = [['Intercept', intercept], ['Coefficient', coefficient], ['R2', R2], ['MAE', MAE], ['MSE', MSE], ['RMSE', RMSE]]
print(tabulate(table, headers = ['Model Attribute']))
```

```python
#clustering analysis

#import the sub-module
from sklearn.cluster import KMeans

#based on highly correlated variables, we can cluster according to stage, investments, no_investors, exits, no_funds, drypowder
Xcluster = investor4[["stage", "investments", "no_funds", "drypowder", "exits", "no_investors"]]

Xcluster.info()
```

```python
#getting rid of entries with null values: 524 entries had no null values

Xcluster2 = Xcluster.dropna()
Xcluster2.info()
```

```python
#scaling variables

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

Xcluster_scaled = scaler.fit_transform(Xcluster2)
```

```python
# instantiate the model with 4 clusters (NB I tried cluster numbers 3-8 and found 4 to have the best distribution)
km = KMeans(n_clusters=4, random_state=1)

```

```python
#fit the model
km.fit(Xcluster_scaled)
```

```python
#labelling our data with cluster numbers and storing it in a separate column

Xcluster2["cluster_number"] = km.labels_

```

```python
Xcluster2.head(20)
```

```python
#finding how many funds fall into each cluster and checking it is well distributed. 
#NB I ran a number of cluster numbers and 4 produced the most evenly distributed.
Xcluster2["cluster_number"].value_counts()
```

```python
#finding the cluster centres (mean average for our variables according to each cluster)
Xcluster2.groupby(["cluster_number"]).mean()

```
